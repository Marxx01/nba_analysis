```python
import pandas as pd

df = pd.read_csv("./data/predecir_mvp.csv", delimiter=";")
```

```python
df
```

```python
years = list(set(df["Year"]))

year_train = list(range(1984, 2021))
year_valid = [2021]
year_test = [2022]

df_train = df[df["Year"].isin(year_train)]
df_valid = df[df["Year"].isin(year_valid)]
df_test = df[df["Year"].isin(year_test)]
df_train.reset_index(drop=True, inplace=True)
df_valid.reset_index(drop=True, inplace=True)
df_test.reset_index(drop=True, inplace=True)


len(df), len(df_train), len(df_valid), len(df_test)
```

```python
def seleccionar_columnas_entrenar(df):
    train_columns = [
        'Pos', 'Age', 'G', 'GS', 'MP', 'FG', 'FGA', 'FG%', '3P',
        '3PA', '3P%', '2P', '2PA', '2P%', 'eFG%', 'FT', 'FTA', 'FT%', 'ORB',
        'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS', 'Year',
        'W', 'L', 'Finish', 'T-Age',  'T-G', 'T-MP', 'T-FG',
        'T-FGA', 'T-FG%', 'T-3P', 'T-3PA', 'T-3P%', 'T-2P', 'T-2PA', 'T-2P%',
        'T-FT', 'T-FTA', 'T-FT%', 'T-ORB', 'T-DRB', 'T-TRB', 'T-AST', 'T-STL',
        'T-BLK', 'T-TOV', 'T-PF', 'T-PTS', 
    ]
    return df[train_columns]


def seleccionar_columna_correcta(df):   
    return list(df['Player']), df['Share']
```

```python
X = seleccionar_columnas_entrenar(df_train)
nombres, Y_share = seleccionar_columna_correcta(df_train)
```

```python
X
```

```python
Y_share
```

```python
nombres
```

## Un modelo para el porcentaje de votos

```python
from sklearn import svm
from sklearn.metrics import mean_squared_error
```

```python
model_share = svm.SVR()
model_share.fit(X, Y_share)
```

## Predecir y evaluar temporada 2021-2022

```python
X_para_predecir = seleccionar_columnas_entrenar(df_valid)
jugadores, Y_correcta = seleccionar_columna_correcta(df_valid)
```

```python
Y_predict = model_share.predict(X_para_predecir)
```

```python
Y_correcta
```

```python
Y_predict
```

```python
len(Y_predict), len(Y_correcta)
```

```python
mse = mean_squared_error(Y_correcta, Y_predict)
print("MSE:", mse)
```

```python
Y_predict = [0 if y < 0.01 else y for y in Y_predict]
mse = mean_squared_error(Y_correcta, Y_predict)
print("MSE:", mse)
```

```python
Y_predict
```

```python
sum(Y_predict)
```

```python
final = list(zip(Y_predict, jugadores))
```

```python
len(jugadores), len(Y_predict), len(final)
```

```python
list(reversed(list(sorted(final))))[:15]
```

# El modelo ha acertado el MVP 2021-2022

# Nikola Jokić
