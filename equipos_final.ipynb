```python
import pandas as pd
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display
from ipyevents import Event

#DataSet
df = pd.read_csv('./data/NbaFullTeamsLimpia.csv', delimiter = ';', index_col = 0)
df1 = pd.read_csv('./data/teamID.csv', delimiter = ';', index_col = 0)

#Variables de la NBA
teams = ['BOS', 'MIL','PHI', 'CLE', 'NOH', 'MIA', 'NYK', 'ATL', 'CHI', 'TOR', 'WAS','IND', 'ORL', 'CHA', 'DET','DEN', 
         'MEM', 'SAC', 'DAL', 'PHO', 'LAC', 'NJN', 'MIN', 'GSW', 'POR', 'UTA', 'OKC', 'LAL', 'SAS', 'HOU']

seasons = [f'{y}-{str(y+1)[2:]}' for y in range(1939, 2024)]
years = list(range(1946,2023))

#Nba Api Equipos Inf
teamID = pd.read_csv('./data/teamID.csv', delimiter = ';')

#Aux
dic_team = {}
dic_year = {}

#Dic con los equipos
for team in teams:
    df_team = df.loc[df['Team'] == team]
    df_ordenado = df_team.sort_values(by = 'Season', ascending = True)
    dic_team[team] = df_ordenado

for year in years:
    df_team = df.loc[df['Year'] == year]
    df_ordenado = df_team.sort_values(by = 'Year', ascending = True)
    dic_year[year] = df_ordenado



def teams3(Funciones):
    def Comparación_Anual(Año, Estadística):
        eq = dic_year[Año]
    
        fig, ax = plt.subplots(figsize=(20,10))
        ax.grid(True)
        ax.set_title(f'La estadística: {Estadística} del año: {Año}')
        #ax.set_xticks(range(1940, 2026, 5)) 
        ax.bar(eq['Team'], eq[Estadística]/eq['G'])

    def Evolución_del_Equipo(Filtro, Estadística):
        if Filtro == 'Equipos':
            def print_team(Equipo):
                fig, ax = plt.subplots(figsize=(20,10))
                ax.grid(True)
                ax.set_title(f'{Estadística} de {Equipo}')
                ax.set_xticks(range(1940, 2026, 5))
                tm = dic_team[Equipo]
                ax.plot(tm['Year'], tm[Estadística]/tm['G'], label = Equipo)
                ax.legend(fancybox=True, framealpha=1, shadow=True, borderpad=1)
            widgets.interact(print_team, Equipo = ['BOS', 'MIL','PHI', 'CLE', 'NJN', 'MIA', 'NYK', 'ATL', 'CHI',
                                                   'TOR', 'WAS','IND', 'ORL', 'CHA', 'DET','DEN', 'MEM', 'SAC', 'DAL',
                                                   'PHO', 'LAC', 'NOH', 'MIN', 'GSW', 'POR', 'UTA', 'OKC', 'LAL',
                                                   'SAS', 'HOU']);
            
        elif Filtro == 'Conferencia':
            dic_conf = {'Este': ['BOS', 'MIL','PHI', 'CLE', 'NJN', 'MIA', 'NYK', 'ATL', 'CHI', 'TOR', 'WAS','IND',
                                 'ORL', 'CHA', 'DET'], 
                        'Oeste': ['DEN', 'MEM', 'SAC', 'DAL', 'PHO', 'LAC', 'NOH', 'MIN', 'GSW', 'POR', 'UTA', 'OKC',
                                  'LAL', 'SAS', 'HOU']}
            def print_conf(Conferencia):
                fig, ax = plt.subplots(figsize=(20,10))
                ax.grid(True)
                ax.set_title(f'{Estadística} de le conferencia: {Conferencia}')
                ax.set_xticks(range(1940, 2026, 5))
                for team in dic_conf[Conferencia]:
                    tm = dic_team[team]
                    ax.plot(tm['Year'], tm[Estadística]/tm['G'], label = team)
                ax.legend(fancybox=True, framealpha=1, shadow=True, borderpad=1)
            widgets.interact(print_conf, Conferencia = ['Este', 'Oeste']);
                
        elif Filtro == 'División':
            dic_div = {'Atlántico': ['TOR', 'NJN', 'NYK', 'BOS', 'PHI'], 
                       'Central': ['IND', 'CHI', 'CLE', 'DET', 'MIL'], 
                       'SurEste':['MIA', 'WAS', 'CHA', 'ATL', 'ORL'],
                    'Pacífico': ['LAC', 'GSW', 'PHO', 'SAC', 'LAL'], 
                       'SurOeste': ['SAS', 'HOU', 'MEM', 'DAL', 'NOH'], 
                       'NorOeste': ['DEN', 'MIN', 'POR', 'UTA', 'OKC']}
            def print_div(División):
                fig, ax = plt.subplots(figsize=(20,10))
                ax.grid(True)
                ax.set_title(f'{Estadística} de la división: {División}')
                ax.set_xticks(range(1940, 2026, 5))
                for team in dic_div[División]:
                    tm = dic_team[team]
                    ax.plot(tm['Year'], tm[Estadística]/tm['G'], label = team)
                ax.legend(fancybox=True, framealpha=1, shadow=True, borderpad=1)
            widgets.interact(print_div, División = ['Atlántico', 'Central', 'SurEste', 'Pacífico', 'SurOeste', 
                                                    'NorOeste']);
        
        elif Filtro == 'Todos':
            teams = ['BOS', 'MIL','PHI', 'CLE', 'NOH', 'MIA', 'NYK', 'ATL', 'CHI', 'TOR', 'WAS','IND', 'ORL', 'CHA',
                     'DET','DEN', 'MEM', 'SAC', 'DAL', 'PHO', 'LAC', 'NJN', 'MIN', 'GSW', 'POR', 'UTA', 'OKC', 'LAL',
                     'SAS', 'HOU']
            fig, ax = plt.subplots(figsize=(20,10))
            ax.grid(True)
            ax.set_title(f'{Estadística} de todos los equipos de la NBA')
            ax.set_xticks(range(1940, 2026, 5))
            for team in teams:
                tm = dic_team[team]
                ax.plot(tm['Year'], tm[Estadística]/tm['G'], label = team)
            ax.legend(fancybox=True, framealpha=1, shadow=True, borderpad=1)

    def Localizador_de_Equipos(txt):
        teams = df1
        l = []
        for fila in teamID.iterrows():
            fila_l = list(fila)
            team = fila[1][2]
            ab = fila_l[1][3]
            if team.lower().find(txt.lower()) != -1:
                if team not in l:
                    l.append([team, ab])
                        
        if len(l) == 0:
            return 'No hay equipos que contengas esa porción de texto.'
        else:
            print(l)
            
    if Funciones == 'Comparación_Anual':
        widgets.interact(
            Comparación_Anual,
            Año = widgets.Play(min = 1949, max = 2022), 
            Estadística = [
                'W', 'L', 'Finish', 'G', 'MP', 'FG', 'FGA', 'FG%','3P', '3PA',
                '2P', '2PA', 'FT', 'FTA', 'ORB', 'DRB', 'TRB', 'AST', 'STL',
                'BLK', 'TOV', 'PF', 'PTS'
            ],
            freq=(1, 5, 0.5)
        );
    
    elif Funciones == 'Localizador_de_Equipos':
        def on_text_change(change):
            with out:
                out.clear_output()
                result = Localizador_de_Equipos(change['new'])

        txt = widgets.Text(
            value='',
            placeholder ='Ingrese el nombre del Equipo',
            description = 'Buscador:',
            disabled=False)
            
            
        out = widgets.Output()
        display(txt, out)
        txt.observe(on_text_change, names='value')
            
    elif Funciones == 'Evolución_del_Equipo': 
        widgets.interact(
            Evolución_del_Equipo, 
            Filtro = ['-','Equipos', 'Conferencia', 'División', 'Todos'], 
            Estadística = ['W', 'L', 'Finish', 'G', 'MP','FG', 'FGA', 'FG%', '3P', '3PA', '2P', '2PA', 'FT', 
                           'FTA', 'ORB', 'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS'] 
        );
        
        
widgets.interact(teams3, Funciones  = ['-','Localizador_de_Equipos', 'Comparación_Anual', 'Evolución_del_Equipo']);
```
