# Cálculo de correlaciones

## Importamos los DataSets

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

mvps = pd.read_csv('./data/mvps.csv', delimiter = ';', index_col = 0)
dpoy = pd.read_csv('./data/dpoy.csv', delimiter = ';', index_col = 0)
roy = pd.read_csv('./data/roy.csv', delimiter = ';', index_col = 0)
smoy = pd.read_csv('./data/smoy.csv', delimiter = ';', index_col = 0)
```

```python
del[mvps['Rank']]
del[mvps['Player']]
del[mvps['Age']]
del[mvps['Tm']]
del[mvps['Year']]
del[mvps['First']]
del[mvps['Pts Max']]
```

```python
del[dpoy['Rank']]
del[dpoy['Player']]
del[dpoy['Age']]
del[dpoy['Tm']]
del[dpoy['First']]
del[dpoy['Pts Max']]
del[dpoy['Unnamed: 24']]
```

```python
del[roy['Rank']]
del[roy['Player']]
del[roy['Age']]
del[roy['Tm']]
del[roy['First']]
del[roy['Pts Max']]
del[roy['Unnamed: 21']]
```

```python
del[smoy['Rank']]
del[smoy['Player']]
del[smoy['Age']]
del[smoy['Tm']]
del[smoy['First']]
del[smoy['Pts Max']]
del[smoy['Unnamed: 21']]
```

## Obtenemos las correlaciones

```python
smoyC = smoy.corr()
mvpsC = mvps.corr()
royC = roy.corr()
dpoyC = dpoy.corr()
```

## Mostramos las gráficas

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10, 10))
sns.heatmap(
    smoyC,
    annot     = True,
    cbar      = False,
    annot_kws = {"size": 8},
    vmin      = -1,
    vmax      = 1,
    center    = 0,
    cmap      = sns.diverging_palette(20, 220, n=200),
    square    = True,
    ax        = ax,
    fmt = '.2f'
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation = 45,
    horizontalalignment = 'right',
)
ax.set_title('Matriz de Correlación de SMOY')
ax.tick_params(labelsize = 10)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10, 10))
sns.heatmap(
    mvpsC,
    annot     = True,
    cbar      = False,
    annot_kws = {"size": 8},
    vmin      = -1,
    vmax      = 1,
    center    = 0,
    cmap      = sns.diverging_palette(20, 220, n=200),
    square    = True,
    ax        = ax,
    fmt = '.2f'
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation = 45,
    horizontalalignment = 'right',
)
ax.set_title('Matriz de Correlación de MVPs')
ax.tick_params(labelsize = 10)
```
