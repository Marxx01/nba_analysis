## WebScraping

Recomendación de no ejecutar el siguiente código, ya que, empezará con las descargas del html. Para que no haga falta ejecutar esto también hemos subido el html obtenido. Además de que no funcionará ya que es necesario descargar selenium y chromedriver para hacer el webscraping bien.

```python
import os
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import requests

years = list(range(1950,2024))

# install selenium chrome driver from https://chromedriver.chromium.org/downloads
# xattr -d com.apple.quarantine chromedriver
driver = webdriver.Chrome(executable_path="/Users/mhurben/upv/chromedriver")

for year in years:
    url = f'https://www.basketball-reference.com/leagues/NBA_{year}_totals.html#totals_stats'
    
    driver.get(url)
    driver.execute_script("window.scrollTo(1,10000)")
    time.sleep(5)
    
    with open("./html/jugadores/{}.html".format(year), "w+") as f:
        f.write(driver.page_source)
```

## Obtención del dataset

```python
from bs4 import BeautifulSoup
import pandas as pd

years = list(range(1950,2024))

dfs = []
for year in years:
    with open("./html/jugadores/{}.html".format(year)) as f:
        page = f.read()
    
    soup = BeautifulSoup(page, 'html.parser')
    #soup.find('tr', class_="thead").decompose()
    player_table = soup.find_all(id="div_totals_stats")[0]
    player_df = pd.read_html(str(player_table))[0]
    player_df["Year"] = str(year)
    dfs.append(player_df)
    
nba_players = pd.concat(dfs)
```
