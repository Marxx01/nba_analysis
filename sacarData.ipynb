## Obtención del DataSet

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import time

teams = ['BOS', 'MIL','PHI', 'CLE', 'MIA', 'NYK', 'ATL', 'CHI', 'TOR', 'WAS','IND', 'ORL', 'CHA', 'DET','DEN', 
         'MEM', 'SAC', 'DAL', 'PHO', 'LAC', 'MIN', 'GSW', 'POR', 'UTA', 'OKC', 'LAL', 'SAS', 'HOU', 'NJN', 'NOH']

df = []
#teams = ['NJN', 'NOH']

for team in teams:
    #print(team)
    with open(f'./html/dataframeEquipos/{team}.html') as f:
            page = f.read()
    soup = BeautifulSoup(page, 'html.parser')
    table = soup.find(id = 'stats')
    nba_stats = pd.read_html(str(table))[0]
    nba_stats['Team'] = team
    nba_stats['Year'] = nba_stats['Season'].str.slice(0,4)
    df.append(nba_stats)
    
    
nba_df = pd.concat(df)
```
