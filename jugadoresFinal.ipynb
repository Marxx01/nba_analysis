```python
import pandas as pd
import ipywidgets as widgets
from IPython.display import display
import matplotlib.pyplot as plt

#Datasets
df = pd.read_csv('./data/JugadoresNBA.csv', delimiter = ';')
df1 = pd.read_csv('./data/playersID.csv', delimiter = ';')
gameslog = pd.read_csv('./data/_gamelogs.csv', delimiter = ';', index_col=0)

#Variables
teams = ['BOS', 'MIL','PHI', 'CLE', 'NJN', 'MIA', 'NYK', 'ATL', 'CHI', 'TOR', 'WAS','IND', 'ORL', 'CHA', 'DET','DEN', 
         'MEM', 'SAC', 'DAL', 'PHO', 'LAC', 'NJN', 'MIN', 'GSW', 'POR', 'UTA', 'OKC', 'LAL', 'SAS', 'HOU']

#Aux
dic_team = {}

#creador de dic_team
for team in teams:
    df_team = df.loc[df['Tm'] == team]
    df_ordenado = df_team.sort_values(by = 'Year', ascending = True)
    dic_team[team] = df_ordenado


# Función para obtener la evolución de un jugador en una estadística específica
def player_evo(Jugadores, Estadística):
    df_player = stats_players(Jugadores)

    fig, ax = plt.subplots(figsize=(20,10))
    ax.grid(True)
    ax.set_title(f'La evolución de la estadística: {Estadística} para {Jugadores}')
    ax.set_xticks(range(1940, 2026, 5))
    ax.plot(df_player['Year'], df_player[Estadística]/df_player['G'], label=Jugadores)
    ax.legend(fancybox=True, framealpha=1, shadow=True, borderpad=1)

# Función para obtener los registros de partidos de un jugador en una estadística específica
def partidos_jugador(Jugador, Estadística):
    #from nba_api.stats.static import players
    #from nba_api.stats.endpoints import playergamelog
    #from nba_api.stats.library.parameters import SeasonAll
    import pandas as pd
    
    pid = df1.loc[df1['full_name'] == Jugador].iloc[0]['id']
    gl = gameslog.loc[gameslog['Player_ID'] == pid].sort_values(by=Estadística, ascending=False)

    return gl

# Función para comparar la evolución de varios jugadores en una estadística específica
def comp(Jugadores1, Jugadores2, Estadística):
    fig, ax = plt.subplots(figsize=(20,10))
    ax.grid(True)
    ax.set_title(f'Comparación de {Jugadores1} y {Jugadores2} mediante la Estadística: {Estadística}')
    ax.set_xticks(range(1940, 2026, 5))
    df1 = df.loc[df['Player'] == Jugadores1]
    ax.plot(df1['Year'], df1[Estadística]/df1['G'], label=Jugadores1)
    df2 = df.loc[df['Player'] == Jugadores2]
    ax.plot(df2['Year'], df2[Estadística]/df2['G'], label=Jugadores2)
    ax.legend(fancybox=True, framealpha=1, shadow=True, borderpad=1)

def loc_jugadores(txt, retirado):
    activos = []
    retirados = []
    p = df1
    for fila in df1.iterrows():
        fila_l = list(fila)
        player = fila[1][2]
        if fila[1][5]:
            if txt.lower() in player.lower():
                activos.append(player)
        else:
            if txt.lower() in player.lower():
                retirados.append(player)
    if retirado:
        return retirados
    else:
        return activos

def stats_players(p):
    return df.loc[df['Player'] == p]

def ejecutar_accion(Funciones):
    if Funciones == 'Evolución_del_Jugador':
        @widgets.interact
        def manejar_jugador(Jugador='', Retirado=False):
            jugadores = loc_jugadores(Jugador, Retirado)
            widgets.interact(
                player_evo,
                Jugadores= jugadores,
                Estadística = ['FG', 'FGA', 'FG%', '3P', '3PA', '2P', 
                        '2PA', 'FT', 'FTA', 'ORB', 'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS']
            );
    elif Funciones == 'Mostrar_Partidos':
        @widgets.interact
        def manejar_jugador(Jugador='', Retirado=False):
            jugadores = loc_jugadores(Jugador, Retirado)
            widgets.interact(
                partidos_jugador,
                Jugador = jugadores, 
                Estadística = ['MIN', 'FGM', 'FGA', 'FG_PCT', 'FG3M', 'FG3A', 'FG3_PCT', 'FTM', 'FTA', 'FT_PCT', 
                        'OREB', 'DREB', 'REB', 'AST', 'STL', 'BLK', 'TOV', 'PF','PTS', 'PLUS_MINUS']
            );
    elif Funciones == 'Comparar_dos_Jugadores':
        @widgets.interact
        def manejar_jugador(Jugador1='', Jugador2='', Retirado1=False, Retirado2=False):
            jugadores1 = loc_jugadores(Jugador1, Retirado1)
            jugadores2 = loc_jugadores(Jugador2, Retirado2)
            widgets.interact(
                    comp,
                    Jugadores1 = list(jugadores1),
                    Jugadores2 = list(jugadores2),
                    Estadística = ['FG', 'FGA', 'FG%', '3P', '3PA', '2P', 
                            '2PA', 'FT', 'FTA', 'ORB', 'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS']
            )
            
        
            
            
widgets.interact(
    ejecutar_accion, 
    Funciones = ['-','Evolución_del_Jugador', 'Mostrar_Partidos', 'Comparar_dos_Jugadores']
)
```
